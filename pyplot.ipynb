2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a sequence of random numbers
np.random.seed(0)  # For reproducibility
random_numbers = np.random.randint(0, 100, 5)  # Generate 5 random integers between 0 and 100

# Create figure and axes objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(range(len(random_numbers)), random_numbers, color='blue')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')

# Plot
ax2.plot(range(len(random_numbers)), random_numbers, color='red', marker='o')
ax2.set_title('Line Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')

# Show the plot
plt.tight_layout()
plt.show()
```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility
random_numbers = np.random.randint(0, 100, 5)  # Generate 5 random integers between 0 and 100

# Create figure and axes objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot with different style and color
ax1.scatter(range(len(random_numbers)), random_numbers, color='green', marker='x')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')

# Plot with different style and color
ax2.plot(range(len(random_numbers)), random_numbers, color='orange', marker='^', linestyle='--')
ax2.set_title('Line Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')

# Show the plot
plt.tight_layout()
plt.show()
```
